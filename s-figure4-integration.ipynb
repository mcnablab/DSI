```python
import numpy as np
import nibabel as nib
import dsi4sampling as dsi
from dipy.viz import fvtk
from dipy.data import get_sphere
import dsisamplingtool as dst

%pylab inline
np.set_printoptions(threshold=numpy.nan)
```

```python
## Load ODF sphere directions
sphere = get_sphere('symmetric724')
```

```python
files = np.array(['DSI11_exvivo_cc', 'DSI11_exvivo_xfib', 'DSI11_invivo_cc', 'DSI11_invivo_xfib'])
```

```python
qgridsz = 201;
pdfweight = 2;

for ii in np.arange(files.shape[0]):
    filename = files[ii]
    data = dst.loaddata(filename);
    gtab = dst.loadgtab(filename);
    fov, mdd = dst.loadstat(filename);
    pdfroi, pdfslicestart, pdfsliceend, camerapos = dst.loadroi(filename, qgridsz);
    
    widths = np.array([np.inf, 4. *dsi.create_qtable(gtab).max()])
    
    for hannwidth in widths: # no hanning filter, hanning filter with 0 roll-off weighting
        
        distances = np.array([mdd, fov/2]);
        
        for integdistance in distances:
            rend = np.round(integdistance / fov * qgridsz);
            dsimodel = dsi.DiffusionSpectrumModel(gtab, qgrid_size=qgridsz, filter_width=hannwidth, r_start=0, r_end=rend, r_step=0.2);
            dsifit = dsimodel.fit(data);
            odf = dsifit.odf(sphere, pdfweight);
            pdf = dsifit.pdf();
            
            # save odf
            r = fvtk.ren();
            fvtk.add(r, fvtk.sphere_funcs(odf, sphere));
            fvtk.camera(r, pos=(camerapos[0], camerapos[1], camerapos[2]));
            odfname = 'figure/figure4/' + filename + '_odf_hann_' + str(hannwidth) + '_integdist_' + str(integdistance) + '.png';
            fvtk.record(r, n_frames=1, out_path=odfname, size=(600,600))
            #fvtk.show(r)
            
            # save pdf
            pdfslice = pdf[pdfroi[0], pdfroi[1], pdfroi[2]][pdfslicestart[0]:pdfsliceend[0], pdfslicestart[1]:pdfsliceend[1], pdfslicestart[2]:pdfsliceend[2]];
            pdfslice = np.squeeze(pdfslice);
            fig = figure()
            imshow(pdfslice);
            circle = Circle((qgridsz//2, qgridsz//2), radius = rend, color = 'w', fill=False, linestyle = 'solid', linewidth=3.)
            fig.gca().add_artist(circle)
            axis('off');
            pdfname = 'figure/figure4/' + filename + '_pdf_hann_' + str(hannwidth) + '_integdist_' + str(integdistance) + '.png';
            savefig(pdfname, dpi = 200)
            
            # save weighted pdf
            weightedpdfslice = dst.weightpdfslice(pdfslice); 
            fig = figure()
            imshow(weightedpdfslice);
            circle = Circle((qgridsz//2, qgridsz//2), radius = rend, color = 'w', fill=False, linestyle = 'solid', linewidth=3.)
            fig.gca().add_artist(circle)
            axis('off');
            pdfname = 'figure/figure4/' + filename + '_weightedpdf_hann_' + str(hannwidth) + '_integdist_' + str(integdistance) + '.png';
            savefig(pdfname, dpi = 200)
            
```
