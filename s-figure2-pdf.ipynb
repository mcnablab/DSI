```python
import numpy as np
import nibabel as nib
import dsiadapt as dsi
from dipy.viz import fvtk
from dipy.data import get_sphere
import dsitool as dst
import dipy.reconst.dti as dti
import dipy.core.gradients as grad

%pylab inline
np.set_printoptions(threshold=numpy.nan)
```

```python
# Four datasets
files = np.array(['DSI11_invivo_cc', 'DSI11_exvivo_cc', 'DSI15_exvivo_cc', 'DSI17_exvivo_cc'])
```

```python
# Estimate diffusion rates using DTI fitting
rd = np.zeros(files.shape[0])
mdd = np.zeros(files.shape[0])
for ii in np.arange(files.shape[0]):
    filename = files[ii]
    data = dst.loaddata(filename);
    gtab = dst.loadgtab(filename);
    
    mask = mask3000 = gtab.bvals <= 3000; # Use only data with bvals less than 3000
    datamask = data[:, :, :, mask]
    gtabmask = grad.gradient_table_from_bvals_bvecs(gtab.bvals[mask], gtab.bvecs[mask, :])
    disp(gtabmask.bvals.shape) # Number of measurements used for DTI fitting
    
    tensormodel = dti.TensorModel(gtabmask)
    tensorfit = tensormodel.fit(datamask)
    rd[ii] = tensorfit.ad.mean()
    
    bigdelta, smalldelta = dst.loadtime(filename);
    mdd[ii] = np.sqrt(6 * rd[ii] * (bigdelta - smalldelta / 3))
```

```python
# Diffusion rate in the principle orientation in mm2/s
rd[0], rd[1], rd[2], rd[3]
```

```python
# Mean displacement distnace in the principle orientation. Unit: mm
mdd[0], mdd[1], mdd[2], mdd[3]
```

```python
# For ex vivo, use the estimation from DSI17 since it has the most DWIs among the three datasets
mdd[1] = mdd[3]
mdd[2] = mdd[3]
```

```python
# Compute field-of-view (FOV)
fov = np.zeros(files.shape[0])
for ii in np.arange(files.shape[0]):
    filename = files[ii]
    data = dst.loaddata(filename);
    gtab = dst.loadgtab(filename);
    bigdelta, smalldelta = dst.loadtime(filename);
    
    bmax = gtab.bvals.max(); # Unit: s/mm2
    qmax = np.sqrt(bmax / (bigdelta - smalldelta / 3)) / (2 * np.pi); # Unit:mm-1
    deltaq = qmax / dsi.create_qtable(gtab).max(); # Unit: mm-1
    fov[ii] = 1 / deltaq; # Unit: mm
```

```python
fov[0], fov[1], fov[2], fov[3]
```

```python
np.save('results/fov', fov);
np.save('results/mdd', mdd);
```

```python
qgridsz = 201; # Q-space grid size with extensive zero padding for better visualization
for ii in np.arange(files.shape[0]):
    filename = files[ii]
    data = dst.loaddata(filename);
    gtab = dst.loadgtab(filename);

    hannwidth = np.inf;
    integdist = mdd[ii];
    rend = np.round(integdist / fov[ii] * qgridsz);
    dsimodel = dsi.DiffusionSpectrumModel(gtab, qgrid_size=qgridsz, filter_width=hannwidth, r_start=0, r_end=rend, r_step=0.2);
    dsifit = dsimodel.fit(data);
    pdf = dsifit.pdf(); # The pdf is normalized such that the summation is 1
    
    pdfvox = pdf[3, 0, 0] # Show one voxel
    pdfslice = np.rot90(pdfvox[:, :, qgridsz//2 + 1])
    pdfprofile = pdfslice[:, qgridsz//2 + 1]
    
    figure();
    imshow(pdfslice);
    circle = Circle((qgridsz//2, qgridsz//2), radius = rend, color = 'r', fill=False, linestyle = 'dashed', linewidth=3.)
    gca().add_artist(circle)
    axis('off');
    colorbar();
    pdfname = 'figure/figure2/' + filename + '_pdf.png';
    #savefig(pdfname, dpi = 600)
```
