```python
import numpy as np
import nibabel as nib
import numpy.matlib as matlab
import dipy.core.gradients as grad
import dipy.reconst.dsi as dsi
    
%pylab inline
np.set_printoptions(threshold=numpy.nan)
```

```python
def squarepdf(pdfslice):
    gridsize = pdfslice.shape[0]
    center = gridsize//2
    tmp = np.arange(gridsize) - center
    x, y = meshgrid(tmp, tmp)
    r2 = x ** 2 + y ** 2
    squarepdfslice = pdfslice * r2 / float((gridsize ** 2))
    return squarepdfslice
```

```python
# Data path
data_path = '/biac4/wandell/data/qytian/DSIProject/'
```

```python
# Load data

# Load DSI515 data
DSI515_bx = nib.load(data_path + 'DSI515/DSI515_bx_reg_cc.nii.gz').get_data()
DSI515_b0 = nib.load(data_path + 'DSI515/DSI515_b0_reg_cc.nii.gz').get_data().mean(-1)
DSI515 = np.concatenate([DSI515_b0[..., None], DSI515_bx], axis=-1)

# Load DSI1418 data
DSI1418_bx = nib.load(data_path + 'DSI1418/DSI1418_bx_reg_cc.nii.gz').get_data()
DSI1418_b0 = nib.load(data_path + 'DSI1418/DSI1418_b0_reg_cc.nii.gz').get_data().mean(-1)
DSI1418 = np.concatenate([DSI1418_b0[..., None], DSI1418_bx], axis=-1)

# Load DSI2109 data
DSI2109_bx = nib.load(data_path + 'DSI2109/DSI2109_bx_reg_cc.nii.gz').get_data()
DSI2109_b0 = nib.load(data_path + 'DSI2109/DSI2109_b0_reg_cc.nii.gz').get_data().mean(-1)
DSI2109 = np.concatenate([DSI2109_b0[..., None], DSI2109_bx], axis=-1)

# Load DSI515 data
DSI515_invivo = nib.load(data_path + 'DSI515_invivo/data_ROI_cc.nii.gz').get_data()
```

```python
# Load bvals and bvecs
DSI515_gtab = grad.gradient_table(data_path + 'DSI515/bvals_515_standard.txt', data_path + 'DSI515/bvecs_prs_515_standard.txt')
DSI1418_gtab = grad.gradient_table(data_path + 'DSI1418/bvals_1418_standard.txt', data_path + 'DSI1418/bvecs_prs_1418_standard.txt')
DSI2109_gtab = grad.gradient_table(data_path + 'DSI2109/bvals_2109_standard.txt', data_path + 'DSI2109/bvecs_prs_2109_standard.txt')
DSI515_invivo_gtab = grad.gradient_table(data_path + 'DSI515_invivo/bvals', data_path + 'DSI515_invivo/bvecs')
```

```python
# Specify statistics
DSI515_fov = 28.0 #um
DSI1418_fov = 39.2 #um
DSI2109_fov = 44.8 #um
DSI515_invivo_fov = 40.5 #um

DSI515_md = 5 #um
DSI1418_md = 5 #um
DSI2109_md = 5 #um
DSI515_invivo_md = 13.4 #um
```

```python
# Compute PDFs. For better visualization, use many zeros paddings.
# Set the filter width to extremely large to turn it off
DSI515_model = dsi.DiffusionSpectrumModel(DSI515_gtab, qgrid_size=201, filter_width=10000000)
DSI515_fit = DSI515_model.fit(DSI515)

DSI1418_model = dsi.DiffusionSpectrumModel(DSI1418_gtab, qgrid_size=201, filter_width=10000000)
DSI1418_fit = DSI1418_model.fit(DSI1418)

DSI2109_model = dsi.DiffusionSpectrumModel(DSI2109_gtab, qgrid_size=201, filter_width=10000000)
DSI2109_fit = DSI2109_model.fit(DSI2109)

DSI515_invivo_model = dsi.DiffusionSpectrumModel(DSI515_invivo_gtab, qgrid_size=201, filter_width=10000000)
DSI515_invivo_fit = DSI515_invivo_model.fit(DSI515_invivo)
```

```python
DSI515_pdfslice = DSI515_fit[0, 0, 0].pdf()[:, :, DSI515_model.qgrid_size//2];
DSI1418_pdfslice = DSI1418_fit[0, 0, 0].pdf()[:, :, DSI1418_model.qgrid_size//2];
DSI2109_pdfslice = DSI2109_fit[0, 0, 0].pdf()[:, :, DSI2109_model.qgrid_size//2];
DSI515_invivo_pdfslice = DSI515_invivo_fit[0, 0, 0].pdf()[:, :, DSI515_invivo_model.qgrid_size//2].T;

DSI515_pdfslice = squarepdf(DSI515_pdfslice)
DSI1418_pdfslice = squarepdf(DSI1418_pdfslice)
DSI2109_pdfslice = squarepdf(DSI2109_pdfslice)
DSI515_invivo_pdfslice = squarepdf(DSI515_invivo_pdfslice)
```

```python
imshow(DSI515_pdfslice);
circle = Circle((DSI515_pdfslice.shape[0]//2, DSI515_pdfslice.shape[1]//2), radius = DSI515_md / DSI515_fov * DSI515_pdfslice.shape[0], color = 'r',fill=False, linestyle = 'dashed')
#colorbar()
gca().add_artist(circle)
axis('off');
savefig('DSI515_PDF.png', dpi = 200)
```
